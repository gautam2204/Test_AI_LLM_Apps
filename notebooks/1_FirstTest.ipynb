### Instead of using standalone (measure method) can be replaced by evaluate method which can be seen in confident AI

```python
import deepeval
import os
from deepeval import test_case
from deepeval import evaluate
from dotenv import load_dotenv
from openai import api_key
from deepeval.test_case import LLMTestCase
from deepeval.metrics import AnswerRelevancyMetric
import os
load_dotenv()

DEEP_EVAL_API_KEY = os.getenv("DEEV_EVAL_KEY")
deepeval.login(DEEP_EVAL_API_KEY)
import sys
# 1. Get the path to the folder ABOVE 'notebooks'
# This points to: d:\pyhton_projects\Test_AI_LLM_Apps
project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))

# 2. Add that root folder to the path
if project_root not in sys.path:
    sys.path.insert(0, project_root) # insert at 0 to prioritize this path
from modules.local_model_llama import local_judge


# os.environ["OPENAI_API_KEY"] = "your_api_key_here"

answerRelevancyMetric=AnswerRelevancyMetric(model=local_judge,threshold=0.7)

test_case = LLMTestCase(
    input="What is capital of India?",
    actual_output="New Delhi",
    expected_output="New Delhi is the Capital of India."
)

evaluate(test_cases=[test_case],metrics=[answerRelevancyMetric])
```
