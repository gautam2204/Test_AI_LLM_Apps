### Evaluate with Golden and Evaluation Daaset

```python
from deepeval.dataset import Golden,EvaluationDataset
from deepeval.metrics import AnswerRelevancyMetric
import sys
sys.path.append('..')
from modules.local_model_llama import local_judge


```

```python
metric=AnswerRelevancyMetric(model=local_judge)
```

```python


# create golden instead of test cases

golden1 = Golden(
    input="What is the Capital of India?",
    actual_output="New Delhi is Capital of Repubic of India", #This will come from LLM
    expected_output="New Delhi is the Capital of India."
)

golden2 = Golden(
    input="What is the Capital of USA?",
    actual_output="Washington is Capital of USA", #This will come from LLM
    expected_output="Washington."
)


golden1
```

```python
dataSet = EvaluationDataset(goldens=[golden1,golden2])
dataSet
```

```python

from deepeval.test_case import LLMTestCase


for golden in dataSet.goldens:
    test_case = LLMTestCase(
        input=golden.input,
        actual_output=golden.actual_output,
        expected_output=golden.expected_output
    )
    
    dataSet.add_test_case(test_case=test_case)
```

```python
dataSet
```

```python
from deepeval import evaluate
import deepeval
import os

deepeval.login(os.getenv('DEEV_EVAL_KEY'))
evaluate(test_cases=dataSet.test_cases,metrics=[metric])
```
