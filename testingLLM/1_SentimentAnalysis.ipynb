## Sentiment Analysis

```python
from langchain.agents import create_agent
from langchain.messages import HumanMessage, SystemMessage
from langchain_openai import ChatOpenAI


def llm_agent():
    # --- 3. THE AGENT SETUP ---
    model = ChatOpenAI(
        base_url="http://127.0.0.1:8080", 
        api_key="dummy"
    )

    # Use create_agent to build the "Robot"
    # Note: We pass the middleware here so the RAG logic is active!
    return create_agent(
        model=model,
        system_prompt="You are a helpful assistant! Your name is Bob."
        # tools=[search, get_weather]
    )

llm = llm_agent()
response = llm.invoke({"messages": HumanMessage(content="What is your name?")})
print(response)

    
    
```
